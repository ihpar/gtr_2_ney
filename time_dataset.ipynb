```python
import librosa
import numpy as np
from constants import *
from pathlib import Path
import IPython.display as ipd
import matplotlib.pyplot as plt
```

```python
gtr_dir = "dataset/time/gtr"
ney_dir = "dataset/time/ney"

# read all wav file paths from audio dataset dirs

# get all guitar file paths
gtr_files = sorted([f for f in Path(gtr_dir).glob("*.wav")],
                   key=lambda x: int(x.stem.split("_")[1]))

# get all ney file paths
ney_files = sorted([f for f in Path(ney_dir).glob("*.wav")],
                   key=lambda x: int(x.stem.split("_")[1]))
```

```python
# print the paths of the first and the last 2 gtr files
print(len(gtr_files))
print(gtr_files[:2])
print(gtr_files[-2:])

print("-" * 50)

# print the paths of the first and the last 2 ney files
print(len(ney_files))
print(ney_files[:2])
print(ney_files[-2:])
```

```python
# get the signal data of the 100th gtr file
gtr_signal, _ = librosa.load(gtr_files[100], mono=True, sr=SR)
print(gtr_signal.shape)
ipd.Audio(gtr_signal, rate=SR)
```

```python
# get the signal data of the 100th ney file
ney_signal, _ = librosa.load(ney_files[100], mono=True, sr=SR)
print(ney_signal.shape)
ipd.Audio(ney_signal, rate=SR)
```

```python
plt.title("Guitar signal first 200 samples")
plt.scatter(np.linspace(0, 1, 200), gtr_signal[:200])
plt.show()
```

```python
plt.title("Whole guitar signal")
plt.plot(np.linspace(0, 1, len(gtr_signal)), gtr_signal)
plt.show()
```
